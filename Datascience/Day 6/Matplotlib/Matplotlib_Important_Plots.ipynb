<h1 align="center", style="color:green">Matplotlib : Hexbin plot</h1>

**A hexbin plot is a type of two-dimensional histogram that is commonly used to visualize the density of points in a scatter plot. In a hexbin plot, the data is first binned into a grid of hexagons, and then the color of each hexagon represents the number of data points that fall within it. This can be a useful way to visualize dense scatter plots with many overlapping points.**

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate some random data
x = np.random.randn(1000)
y = np.random.randn(1000)

# Create a hexbin plot
plt.hexbin(x, y, gridsize=20, cmap='Blues')
#This parameter represents the number of hexagons in the x-direction or both direction.
# Set the title and axis labels
plt.title('Hexbin Plot')
plt.xlabel('X')
plt.ylabel('Y')

# Add a colorbar
plt.colorbar()

# Display the plot
plt.show()
```

**The Probability Density Function (PDF) is a function that describes the relative likelihood of a continuous random variable taking on a particular value. It gives the probability that the random variable will take on a value within a certain range. The PDF is a non-negative function, and its integral over the entire range of the variable is equal to 1.**

**The Cumulative Distribution Function (CDF) is a function that describes the probability that a continuous random variable is less than or equal to a certain value. It gives the cumulative probability of the random variable taking on a value less than or equal to a certain value.**

**The PDF and CDF are related as follows: the PDF is the derivative of the CDF. In other words, the PDF is the rate of change of the CDF. This means that the CDF can be obtained by integrating the PDF, and the PDF can be obtained by differentiating the CDF.**

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate some random data from a normal distribution
mu, sigma = 0, 0.1
data = np.random.normal(mu, sigma, 1000)

# Plot a histogram with normalized density
plt.hist(data, bins=30, density=True, alpha=0.5, color='green')

# Add a line for the probability density function
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
plt.plot(x, 1/(sigma * np.sqrt(2 * np.pi)) * np.exp(- (x - mu)**2 / (2 * sigma**2)), linewidth=2, color='black')

# Set the title and axis labels
plt.title('Probability Density Plot')
plt.xlabel('Value')
plt.ylabel('Probability Density')

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate some random data from a normal distribution
mu, sigma = 0, 0.1
data = np.random.normal(mu, sigma, 1000)

# Plot a histogram with cumulative normalized density
n, bins, patches = plt.hist(data, bins=30, density=True, cumulative=True, alpha=0.5, color='green')

# Add a line for the true cumulative distribution function
from scipy.stats import norm
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
plt.plot(x, norm.cdf(x, mu, sigma), linewidth=2, color='black')

# Set the title and axis labels
plt.title('Cumulative Distribution Function Plot')
plt.xlabel('Value')
plt.ylabel('Cumulative Probability')

# Show the plot
plt.show()
```

<h1 align="center", style="color:green">Matplotlib : Count Charts</h1>

**A count chart is a type of bar chart that shows the number of occurrences of each category in a dataset. It can be useful for visualizing the distribution of categorical data.**

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate some random categorical data
data = np.random.choice(['A', 'B', 'C', 'D', 'E'], size=100)

# Count the occurrences of each category
counts = np.unique(data, return_counts=True)

# Create a bar chart of the counts
plt.bar(counts[0], counts[1])

# Set the title and axis labels
plt.title('Count Chart')
plt.xlabel('Categories')
plt.ylabel('Counts')

# Display the plot
plt.show()
```

```python
data = np.random.choice(['A', 'B', 'C', 'D', 'E'], size=100)

# Count the occurrences of each category
counts = np.unique(data, return_counts=True)
counts
```

```python
# Generate some random categorical data
data = np.random.choice(['A', 'B', 'C', 'D', 'E'], size=100)
# to generate random samples from a given array or sequence. 
# Count the occurrences of each category
counts = np.unique(data, return_counts=True)

# Define the colors for the bars
colors = ['orange', 'green', 'blue', 'red', 'purple']

# Create a bar chart of the counts with colored bars
plt.bar(counts[0], counts[1], color=colors)

# Set the title and axis labels
plt.title('Count Chart')
plt.xlabel('Categories')
plt.ylabel('Counts')

# Display the plot
plt.show()
```

**The main difference between count charts and histograms is the type of data they are used for. Count charts are used for discrete categorical data, while histograms are used for continuous numerical data. The bars in count charts represent individual categories, while the bars in histograms represent intervals of continuous data.**

**a bar chart is a specific type of chart that uses rectangular bars to represent and compare data in different categories. On the other hand, a count chart, assuming it represents frequencies, can refer to various visual representations (e.g., histogram, frequency polygon) that display the frequency or count of categories or values within a dataset.**

<h1 align="center", style="color:green">Matplotlib : Pie Charts</h1>

<h3 style="color:purple">Draw pie chart to track down home expenses</h3>

```python
exp_vals = [1400,600,300,410,250]
exp_labels = ["Home Rent","Food","Phone/Internet Bill","Car ","Other Utilities"]
plt.pie(exp_vals,labels=exp_labels)
plt.show()
```

<h3 style="color:purple">Draw pie chart as perfect circle</h3>

```python
plt.pie(exp_vals,labels=exp_labels, shadow=True)
plt.axis("equal")
plt.show()
```

<h3 style="color:purple">Show percentages for every pie. Specify radius to increase chart size</h3>

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels, shadow=True, autopct='%1.1f%%',radius=1.5)
plt.show()
```

**Autopct  If you want to label the wedges with their numeric value in a pie chart, then use autopct parameter. This parameter allows us to display the percent value using string formatting. Say for example, the percent value calculated for a segment is 34.678666 and if you want to display the percent value rounded to 1 decimal place then autopct parameter should be assigned the format string %1.1f then the wedge will be labelled with the numeric value 34.6. If you want to add a percent sign (%) to the label then use two percent signs(%%) in the format string so that the special character % escapes itself.**

<h3 style="color:purple">Explode</h3>

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels, shadow=True, autopct='%1.1f%%',radius=1.5,explode=[0,0.5,0.1,0,0.2])
plt.show()
#exp_labels = ["Home Rent","Food","Phone/Internet Bill","Car ","Other Utilities"]
```

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels, shadow=True, autopct='%1.1f%%',radius=1.5,explode=[0,0,0,0.1,0.2],counterclock=True, startangle=45)
plt.show()
```

**Startangle  By default the Startangle is zero, which means starting from the positive x-axis the wedges are arranged in the counter clock wise direction. If you specify a different Startangle then the start of the pie chart is shifted by this angle in degrees and then the wedges are arranged in counter clock wise direction from this position.**

**Counterclock  Specifies the direction in which the wedges are arranged, clockwise or counter clockwise. The default value is True.**

<h2 style="color:green" align="center">Saving chart to a file using savefig</h2>

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels,radius=2, autopct='%1.2f%%', explode=[0,0.1,0.1,0,0])
plt.savefig("piechart1.png", dpi=300, bbox_inches="tight", pad_inches=1, transparent=True)
```

**dpi : [ None | scalar > 0 | 'figure' ]
The resolution in dots per inch. If None, defaults to rcParams["savefig.dpi"] = 'figure'. If 'figure', uses the figure's dpi value.**

**bbox_inches : str or Bbox, optional
Bbox in inches. Only the given portion of the figure is saved. If 'tight', try to figure out the tight bbox of the figure. If None, use savefig.bbox**

**pad_inches : scalar, optional
Amount of padding around the figure when bbox_inches is 'tight'. If None, use savefig.pad_inches**

<h2>You can also save in pdf format and to a different location on your computer by specifying absolute path<h2>
