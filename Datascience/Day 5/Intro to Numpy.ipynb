# Random Functions

**1. random.random() (from Python's built-in random module):
Functionality: Generates a random floating-point number in the range [0.0, 1.0).
Return type: Float.**

                                                                      
**2. random.randint(a, b) (from Python's built-in random module):
Functionality: Generates a random integer between a (inclusive) and b (inclusive).
Return type: Integer.**

**3. numpy.random.randn(d0, d1, ..., dn) (from the NumPy library):
Functionality: Generates an array of random numbers from a standard normal distribution 
(mean 0, standard deviation 1). The shape of the array is specified by d0, d1, ..., dn.
Return type: NumPy array of floats.**

**4. np.random.randint() (NumPy library):
np.random.randint(low, high=None, size=None, dtype=int) generates random integers from a given range.
If only low is provided, it generates a random integer in the range [0, low).
If both low and high are provided, it generates random integers in the range [low, high).
If the size parameter is given, it returns an array of random integers with the specified shape.**

**A slicing operation creates a view on the original array which is just a way of accessing arrays data.Thus the original array
is not copied in the memory.We can check using np.map_share_memory() if two arrays share the same memory block**



```python
import numpy as np
a =np.arange(10)
b=a[::2]
b
```

```python
a
```

```python
#lets change the value of b.
b[3]=10
b
```

```python
np.shares_memory(a,b)
```

```python
c=np.arange(20)
```

```python
np.shares_memory(a,c)
```

```python
It menas they have the same memory allocation. b is subarray of a slice method is called view where it is not copying not move
just sharing the same memory.
```
